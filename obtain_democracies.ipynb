```python
import pandas as pd
import statistics as stats
```

```python
url = 'https://en.wikipedia.org/wiki/Democracy_Index'
webdata = pd.read_html(url)
```

```python
country_scores = webdata[5]
country_scores
```

```python
country_scores['five_year_median'] = country_scores[['2021', '2020', '2019', '2018', '2017']].mean(axis=1)
```

```python
country_scores
```

```python
country_scores.groupby(['Regime type']).count()
```

```python
democracies = country_scores.loc[country_scores['five_year_median'] > 6]
len(set(democracies['Country']))

democracies
```

```python
democracies.groupby(['Regime type']).count()
```
