## Columns Rename

```python
def column_rename(col_name):
    t = col_name.strip()
    t = re.sub('%', 'percent ', t)
    t = re.sub(' ', '_', t)
    t = str.upper(t)
    return t
```

```python
transform_df = df.replace(regex=r'---|\?',value=0)
transform_df = transform_df.drop(columns=['Rank'])
transform_df = transform_df.rename(columns=lambda x: column_rename(x))
transform_df = transform_df.rename(str.upper, axis='columns')

transform_df.columns
```

## Series data conversion & manipulation

```python
transform_df.COUNTRY = transform_df.COUNTRY.replace(regex=r'\(|\)|\*|\d',value='')
```
